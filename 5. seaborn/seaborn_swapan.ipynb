## importing seaborn,matplotlib and other libraries

```python
#if seaborn is not installed try this in the cell(!pip install seaborn)
import seaborn as sns
from matplotlib import pyplot as plt
# or we can write
# import matplotlib.pyplot as plt
# importing another required libraries
import numpy as np
import pandas as pd
```

```python
# importing datasets from seaborn 
# note: you must be connected to internet
fmri=sns.load_dataset("fmri")
fmri.head()
```

## lineplot

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
# grouping data with hue
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.show()
```

```python
#adding styles
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event")
plt.show()
```

```python
# adding markers
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event",markers=True)
plt.show()
```

## barplot

```python
sns.barplot(x="region",y="timepoint",data=fmri)
plt.show()
```

## scatterplot

```python
sns.scatterplot(x="timepoint",y="signal",data=fmri,hue="event",style="event",markers=True)
plt.show()
```

## distplot

```python
sns.distplot(fmri["timepoint"])
plt.show()
```

## jointplot

```python
sns.jointplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

## boxplot

```python
sns.boxplot(x="event",y="signal",data=fmri)
plt.show()
```

## pairplot

```python
sns.pairplot(fmri,hue="event")
plt.show()
```
